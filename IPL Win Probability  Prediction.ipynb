# EDA

```python
total_score_df = delivery.groupby(['matchId', 'innings']).sum()['total_runs'].reset_index()
```

```python
total_score_df.shape
```

```python
total_score_df
```

```python
total_score_df = total_score_df[total_score_df['innings'] ==1]
```

```python
match_df = match.merge(total_score_df[['matchId', 'total_runs']], left_on='matchId', right_on = 'matchId')
```

```python
match_df.head(5)
```

```python
match_df['team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Punjab Kings',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals',
    'Gujarat Titans',
    'Lucknow Super Giants'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Kings XI Punjab','Punjab Kings')
match_df['team2'] = match_df['team2'].str.replace('Kings XI Punjab','Punjab Kings')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df['team1'].unique()
```

```python
match_df.head()
```

```python
match_df.shape
```

```python
match_df['method'].value_counts()
```

```python
match_df = match_df[match_df['method'] != 'D/L']
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
match_df['method'].value_counts()
```

```python
match_df = match_df[['matchId', 'city', 'winner', 'total_runs']]
```

```python
delivery_df = match_df.merge(delivery, on = 'matchId')
```

```python
delivery_df = delivery_df[delivery_df['innings'] ==2]
```

```python
delivery_df.shape
```

```python
delivery_df.head(5)
```

```python
delivery_df['total_runs_y'].dtype
```

```python
delivery_df['current_score'] = delivery_df.groupby('matchId')['total_runs_y'].cumsum()
```

```python
delivery_df.head(5)
```

```python
delivery_df['runs_left'] = (delivery_df['total_runs_x'] - delivery_df['current_score'] + 1)
```

```python
delivery_df.head(5)
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head(5)
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == '0' else '1')
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
delivery_df['player_dismissed'].dtype
```

```python
delivery_df['wickets'] = delivery_df.groupby('matchId')['player_dismissed'].cumsum().values
```

```python
delivery_df.head(5)
```

```python
delivery_df['wickets'] = 10 - delivery_df['wickets']
```

```python
delivery_df.head(50)
```

```python
delivery_df['crr'] = delivery_df['current_score']*6 /(120 - delivery_df['balls_left'])
```

```python
delivery_df.head(5)
```

```python
delivery_df['rrr'] = delivery_df['runs_left']*6 /(delivery_df['balls_left'])
```

```python
delivery_df.head(5)
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis = 1)
```

```python
delivery_df.head(5)
```

```python
final_df = delivery_df[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left','wickets', 'total_runs_x', 'crr', 'rrr', 'result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace = True)
```

```python
final_df.isnull().sum()
```

```python
# X_train.describe()
```

```python
final_df = final_df[final_df['balls_left'] !=0]
```

```python
final_df
```

# model preparation

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size = 0.2, random_state=1)
```

```python
X_train
```

## OHE on string column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse=False, drop = 'first'), ['batting_team', 'bowling_team', 'city'])
]
    , remainder = 'passthrough')
```

### Using LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver = 'liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
X_train.describe()
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
pipe.predict_proba(X_test)[6]
```

### Using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', RandomForestClassifier())
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
pipe.predict_proba(X_test)[6]
```
